```python
import os
import requests
from google.colab import output, userdata
from IPython.display import Javascript
from base64 import b64decode
from groq import Groq
from langchain.document_loaders import PyPDFLoader, OnlinePDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain.chains import RetrievalQA
from langchain_groq import ChatGroq
from deep_translator import GoogleTranslator
from TTS.api import TTS
from concurrent.futures import ThreadPoolExecutor
import json

# Setup Groq API Key
os.environ['GROQ_API_KEY'] = userdata.get('GROQ_API_KEY')

import base64
from IPython.display import display, Javascript
from google.colab import output  # Adjust as needed for non-Colab environments

# JavaScript code for recording audio
RECORD = """
const sleep = time => new Promise(resolve => setTimeout(resolve, time))

const b2text = blob => new Promise(resolve => {
    const reader = new FileReader();
    reader.onloadend = e => resolve(e.srcElement.result);
    reader.readAsDataURL(blob);
})

var record = time => new Promise(async resolve => {
    const stream = await navigator.mediaDevices.getUserMedia({ audio: true });
    const recorder = new MediaRecorder(stream);
    const chunks = [];
    recorder.ondataavailable = e => chunks.push(e.data);
    recorder.start();
    await sleep(time);
    recorder.onstop = async () => {
        const blob = new Blob(chunks);
        const text = await b2text(blob);
        resolve(text);
    };
    recorder.stop();
});
"""

def record(sec=5):
    print("Recording...")
    display(Javascript(RECORD))
    s = output.eval_js('record(%d)' % (sec * 1000))
    print("Done recording!")
    b = base64.b64decode(s.split(',')[1])
    return b

def save_audio(data, filename='recorded_audio.wav'):
    with open(filename, 'wb') as f:
        f.write(data)

# Transcription using Whisper Large V3 Turbo
def transcribe_audio(filename):
    client = Groq()
    with open(filename, "rb") as file:
        transcription = client.audio.transcriptions.create(
            file=(filename, file.read()),
            model="whisper-large-v3-turbo",
            response_format="json"
        )
    return transcription.text

# Language Detection and Translation
def detect_language(text):
    try:
        detected_lang = GoogleTranslator(source='auto', target='en').detect(text)
        return detected_lang
    except Exception as e:
        print(f"Error detecting language: {e}")
        return "en"  # Fallback to English if detection fails

def translate_text(text, src_lang, target_lang='en'):
    try:
        translated_text = GoogleTranslator(source=src_lang, target=target_lang).translate(text)
        return translated_text
    except Exception as e:
        print(f"Error translating text: {e}")
        return text  # Return original text if translation fails

def detect_pdf_language(filepath, is_url=False):
    # Load PDF from URL or local file
    if is_url:
        loader = OnlinePDFLoader(filepath)
    else:
        loader = PyPDFLoader(filepath)

    pages = loader.load_and_split(text_splitter())
    sample_text = pages[0].page_content[:500]
    return detect_language(sample_text), pages

# Parallel Translation Processing
def translate_concurrently(text, src_lang, target_lang='en'):
    with ThreadPoolExecutor() as executor:
        future_translation = executor.submit(translate_text, text, src_lang, target_lang)
        return future_translation.result()

# English Embeddings, Vector Store, and Text Splitter
model_name = "BAAI/bge-small-en-v1.5"  # English embeddings
embeddings = HuggingFaceEmbeddings(model_name=model_name, model_kwargs={"device": "cpu"})
vectorstore = Chroma(embedding_function=embeddings, persist_directory="./chromadb", collection_name="english_docs")

def text_splitter():
    return RecursiveCharacterTextSplitter(chunk_size=512, chunk_overlap=20)

def add_to_vectorstore(pages):
    vectorstore.add_documents(pages)

# RetrievalQA with Adaptive Settings
llm = ChatGroq(model_name="llama-3.1-70b-versatile", temperature=0.1, max_tokens=1000)

def answer_question(question, input_lang):
    num_retrievals = min(10, max(3, len(question.split()) // 5))
    retriever = vectorstore.as_retriever(search_kwargs={"k": num_retrievals})
    qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, return_source_documents=True)
    result = qa.invoke({"query": question})
    answer = result['result']

    # Translate answer to input language, if needed
    if input_lang != "en":
        answer = translate_text(answer, src_lang="en", target_lang=input_lang)

    return answer, retriever._collection.confidence_scores

def summarize_text(text):
    client = Groq()

    if len(text.split()) > 200:
        response = client.chat.completions.create(
            messages=[
                {
                    "role": "system",
                    "content": "You are an assistant that outputs summaries in JSON format."
                               "The JSON structure should be: {\"summary\": \"string\"}"
                },
                {
                    "role": "user",
                    "content": f"Summarize this: {text}"
                }
            ],
            model="gemma2-9b-it",
            temperature=0.1,
            max_tokens=100,
            response_format={"type": "json_object"}
        )

        try:
            summary = json.loads(response.choices[0].message.content)["summary"]
            return summary
        except (json.JSONDecodeError, KeyError) as e:
            print(f"Error parsing JSON response: {e}")
            return "Error: Unable to generate summary."
    else:
        return text

# Text-to-Speech with XTTS-v2 and Voice Cloning
def xtts_v2_speech(text, language, speaker_wav=None):
    tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2", gpu=True)
    tts.tts_to_file(
        text=text,
        file_path="response_audio.wav",
        speaker_wav=speaker_wav,  # Optional, for voice cloning
        language=language
    )
    print("Response saved as 'response_audio.wav'")

# Enhanced Pipeline Function with Text Input Option
def multilingual_pipeline():
    input_option = input("Enter 1 for audio input, 2 for text input: ")

    if input_option == "1":
        # Record and Save Audio
        audio_data = record(5)
        save_audio(audio_data, 'recorded_audio.wav')

        # Transcribe Audio
        transcription = transcribe_audio('recorded_audio.wav')
        print(f"Transcription: {transcription}")

        # Detect Transcription Language
        input_lang = detect_language(transcription)
        query = transcription
    elif input_option == "2":
        query = input("Enter your query: ")
        input_lang = detect_language(query)

    print(f"Input Language: {input_lang}")

    # PDF input option: URL or local file
    pdf_option = input("Enter 1 for URL, 2 for file path: ")
    if pdf_option == "1":
        pdf_url = input("Enter PDF URL: (For demo, use https://arxiv.org/pdf/2211.13580): ")
        pdf_lang, pages = detect_pdf_language(pdf_url, is_url=True)
    elif pdf_option == "2":
        pdf_path = input("Enter PDF file path: ")
        pdf_lang, pages = detect_pdf_language(pdf_path, is_url=False)

    print(f"PDF Language: {pdf_lang}")

    # Align Query with Document Language
    if input_lang != pdf_lang:
        query = translate_concurrently(query, src_lang=input_lang, target_lang=pdf_lang)
        print(f"Aligned Query: {query}")

    # Add Document Chunks to Vector Store
    add_to_vectorstore(pages)

    # Retrieve Answer with Confidence Score
    answer, confidence_scores = answer_question(query, input_lang)
    print(f"Retrieved Answer: {answer}, Confidence Scores: {confidence_scores}")

    # Summarize Answer (if needed)
    answer = summarize_text(answer)
    print(f"Summarized Answer: {answer}")

    # Translate Answer Back (if needed)
    if pdf_lang != input_lang:
        answer = translate_concurrently(answer, src_lang=pdf_lang, target_lang=input_lang)
        print(f"Translated Answer: {answer}")

    # Convert Answer to Speech with XTTS-v2 and Voice Cloning
    xtts_v2_speech(answer, language=input_lang)

# Run the pipeline
multilingual_pipeline()
```
